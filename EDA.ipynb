# Jason 3

```python
J3 = Satellite("./orbital_elements/Jason-3.csv", "./manoeuvres/ja3man.txt")
eda = SatelliteEDA(J3)
eda.col_with_man_trimmed(clip_value=0.1) # Trimmed the extreme values in Jason 3 Brouwer Mean Motion to +- 0.1
eda.col_with_man(start_date = "2018-01-01", end_date = "2019-01-01")

std_b = standardize(J3.df_orbit['Brouwer mean motion'])
sns.boxplot(std_b)
```

```python
# Find outliers
import numpy as np
threshold = 3  
outliers = std_b[np.abs(std_b) > threshold]
outlier_indices = outliers.index
print(outlier_indices)
```
